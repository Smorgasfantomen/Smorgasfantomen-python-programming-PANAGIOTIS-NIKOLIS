### 5a: Dice roll convergence

```python
import numpy
import random

print("100 dice rolls to list")

rollist = []
outlist = []

for n in range(0, 100):
    r = random.randrange(1, 7)
    rollist.append(r)


print(rollist)
print(rollist)


for i in range(len(rollist)):
    if rollist[i] == 6:
        outlist.append(rollist[i])

print(f"{len(outlist)} sixes were rolled.")



```

### 5b: Dice roll convergence

```python
import numpy
import random

print("100 dice rolls to list")

rollist = []
outlist = []

for n in range(0, 100000):
    r = random.randrange(1, 7)
    rollist.append(r)


print(rollist)
print(rollist)


for i in range(len(rollist)):
    if rollist[i] == 6:
        outlist.append(rollist[i])

prob = len(outlist) / len(rollist)

print(f"The probability to roll 6 was {(prob * 100):.2f}%.")



```
