#### II)

Alla ekvationer korsar eller passerar nära origo.
Dock är det tydliga skillnader i koefficienterna, vilket har större effekter ju längre bort punkterna ligger från origo. Ex (-1,8, 1,95) sorteras till det högra klustret med ekvationen f (x) = −0.489x, och h(x) = 800x −120 ger en nästan vertikal linje med väldigt underligt sorteringsutfall.

#### III)

Personligen anser jag att det är svårt att klassificera denna typen av obestämd data, då vi inte har någon information kring vad den representerar, vad vi vill uppnå med sorteringen, och vilka specialfall vi kan stöta på (och vad de kan innebära).

Beroende på klustrens form, storlek, densitet, och interna "subkluster" kanske det inte räcker med en klassificering baserat på en förstagradsekvation heller, då jag upplever detta som ett rätt trubbigt verktyg, hur det än positioneras. Andra-, tredje- och flergradiga ekvationer kan mycket väl vara bättre lämpade för att hantera tidigare nämnda attribut. För min nuvarande nivå är detta dock otroligt abstrakt.

Mitt personliga antagande, baserat på min nuvarande begränsade kunskap är att min egna ekvation (genom medelpunkt && vinkelrät mot medellinje) passar bäst av de fyra alternativen. Kanske kan jag skylla på mitt visuella tankesätt och min tid som CAD-konstruktör, men det kändes direkt intuitivt.

Att bygga den på medelvärden gör den förstås känslig för större avvikelser, och punkterna (ex. ≈(-1,8, 1,95) och ≈(1, -1)) som ligger väldigt nära skiljelinjen hade kanske behövt sorteras in i en tredje kategori för data som är otydlig eller bara gränsfall. För detta hade vi kunnat använda ett linjepar, parallellt och speglat kring den primära klassificeringslinjen, för att dela in data i tydliga kluster, och gränsfall som kräver närmre kontroll.
